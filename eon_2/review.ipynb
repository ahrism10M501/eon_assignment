```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("fab.csv")
```

```python
df.head()
```

```python
df = df.drop("SensorSensorTime", axis=1)
df.head()
```

```python
df.shape
```

```python
df = df.fillna(0)
df.head()
```

```python
df.describe()
```

```python
columns = df.columns
print(columns)
```

```python
ndf = df.to_numpy()
ndf = np.where(ndf < 0, 0 , ndf)
```

```python
mean = np.mean(ndf, axis=0)
std = np.std(ndf, axis=0)
mean.shape
```

```python
def barPlot(data, label):
    plt.figure(figsize=[20, 5])
    plt.bar(columns, data)
    plt.xticks(rotation=90)
    plt.xlabel(label)
    plt.show()
```

```python
barPlot(mean, "mean")
```

```python
barPlot(std, "std")
```
